```python
import spacy
from spacy import displacy

nlp = spacy.load("es_core_news_sm")

frase1 = nlp("Juan come arroz con leche.")
frase2 = nlp("Juan está cansado de esperar.")
frase3 = nlp("Las últimas noticias indican que habrá paro de transportistas en Ecatepec. ")

doc = [frase1, frase2, frase3]

for frase in doc:
    for token in frase:
        print(token.text, token.dep_, token.head.text)
    displacy.serve(doc, style="dep")
```
